```python
#Q23
n = 5

i = 1
while i <= n:
    x = 0
    while x < i:
        print("*", end="")
        x += 1
    print()
    i += 1

i = n - 1
while i > 0:
    x = 0
    while x < i:
        print("*", end="")
        x += 1
    print()
    i -= 1
```
