For Loop

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday" ]
for x in week:
  print(x)
```

Break Statement

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday" ]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print(x)
```

Looping through a string


```python
for x in "week":
  print(x)
```

The range() Function

```python
for x in range(5):
  print(x)
for x in range(2,6):
  print(x)
```

```python
for x in range(2,6,2):
  print(x)
```

Nested Loop

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x,y)
```

While Statement

```python
i = 1
while i < 6:
  print(i)
  i += 1 #i = i+1 Assignment Operator
```

Break statement (While)

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print(i)
    continue
```

The Else Statement


```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
