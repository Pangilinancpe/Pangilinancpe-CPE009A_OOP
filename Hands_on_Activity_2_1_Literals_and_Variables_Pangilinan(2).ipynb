```python
for i in range(1, 4):
    print(f"\n--- Data Entry for Student {i} ---")
    name = input("Enter student name: ")

    # Prelim Section
    phoa = float(input("Enter Prelim Hands-on Activity grade: "))
    pquiz = float(input("Enter Prelim Quiz grade: "))
    pas = float(input("Enter Prelim Assignment grade: "))
    pcs = phoa * 0.5 + pquiz * 0.3 + pas * 0.2
    pex = float(input("Enter Prelim Exam grade: "))
    pg = pex * 0.5 + pcs * 0.5

    # Midterm Section
    mhoa = float(input("Enter Midterm Hands-on Activity grade: "))
    mquiz = float(input("Enter Midterm Quiz grade: "))
    mas = float(input("Enter Midterm Assignment grade: "))
    mcs = mhoa * 0.5 + mquiz * 0.3 + mas * 0.2
    mex = float(input("Enter Midterm Exam grade: "))
    mg = (1/3 * pg) + (2/3 * (mex * 0.5 + mcs * 0.5))

    # Finals Section
    fhoa = float(input("Enter Finals Hands-on Activity grade: "))
    fquiz = float(input("Enter Finals Quiz grade: "))
    fas = float(input("Enter Finals Assignment grade: "))
    fcs = fhoa * 0.5 + fquiz * 0.3 + fas * 0.2
    fex = float(input("Enter Finals Exam grade: "))

    # Final Numerical Grade Calculation
    FG = (1/3 * mg) + (2/3 * (fex * 0.5 + fcs * 0.5))

    # Transmutation Logic
    if FG >= 99:
        transmuted = 1.00
    elif FG >= 96:
        transmuted = 1.25
    elif FG >= 93:
        transmuted = 1.50
    elif FG >= 90:
        transmuted = 1.75
    elif FG >= 87:
        transmuted = 2.00
    elif FG >= 84:
        transmuted = 2.25
    elif FG >= 81:
        transmuted = 2.50
    elif FG >= 78:
        transmuted = 2.75
    elif FG >= 75:
        transmuted = 3.00
    else:
        transmuted = 5.00


    print("-" * 30)
    print(f"Student: {name}")
    print(f"Final Numerical Grade: {FG:.2f}")
    print(f"Transmuted Grade: {transmuted:.2f}")
    print("-" * 30)

print("\nAll 3 student records processed.")
```
