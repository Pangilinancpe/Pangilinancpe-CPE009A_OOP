```python
def compute_term():
    hoa = float(input("Enter Hands-on Activity grade: "))
    quiz = float(input("Enter Quiz grade: "))
    ass = float(input("Enter Assignment grade: "))
    cs = hoa * 0.5 + quiz * 0.3 + ass * 0.2
    exam = float(input("Enter Exam grade: "))
    term_grade = exam * 0.5 + cs * 0.5
    return term_grade

def grade(marks):
    if marks >= 90:
      return "A+"
    elif marks >= 80:
      return "A"
    elif marks >= 70:
      return "B+"
    elif marks >= 60:
      return "B"
    elif marks >= 50:
      return "C"
    elif marks >= 40:
      return "D"
    else:
      return "F"

def remarks(percent):
    if percent >= 90:
      return "Outstanding"
    elif percent >= 80:
      return "Excellent"
    elif percent >= 70:
      return "Very Good"
    elif percent >= 60:
      return "Good"
    elif percent >= 40:
      return "Satisfactory"
    else:
      return "Needs Improvement"

print("=== Student Performance ===")
overall = compute_term()

print("\n=== Report Card ===")
print(f"Overall Grade: {overall:.2f} → {grade(overall)}")
print(f"Performance Remarks: {remarks(overall)}")
```
