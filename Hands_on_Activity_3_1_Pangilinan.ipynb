1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance formula.
coorindates_list = [(1,1), (2,3)]



```python
import math
coordinates_list = [(1, 1), (2, 3)]
p1 = coordinates_list[0]
p2 = coordinates_list[1]
distance = math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

print(f"The distance between {p1} and {p2} is {distance}")
```

2. For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The new
string value should be displayed with the asterisks.


```python
import re

inval = {'Idiot': '******', 'animal':'******' }
initial_string = input("Enter a string: ")
filtered_string = initial_string

for word_to_filter, replacement_pattern in inval.items():
    filtered_string = re.sub(r'\b' + re.escape(word_to_filter) + r'\b',
                             replacement_pattern,
                             filtered_string,
                             flags=re.IGNORECASE)

print("Filtered string:", filtered_string)
```

3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.


```python
phonebook = [
    {'id': '1', 'name': 'Bienvenido Pangilinan Jr.', 'contact': '09060611233', 'address': 'TIP'},
    {'id': '2', 'name': 'Jayne Riane Dejamco', 'contact': '09171234567', 'address': 'URS'}

]

search_id = input("Enter the ID of the record: ")
found = False

for contact in phonebook:
    if contact['id'] == search_id:
        print(f"\nName: {contact['name']}")
        print(f"Contact: {contact['contact']}")
        print(f"Address: {contact['address']}")
        found = True
        break

if not found:
    print("Record not found.")
```
