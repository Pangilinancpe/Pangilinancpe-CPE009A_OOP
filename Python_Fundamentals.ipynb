Comparison Operator

```python
t = 1
u = 2
#Equal
print (t == u)

#Not Equal
print (t != u)

#Greater than
print (t > u)

#Less than
print (t < u)

#Greater than or equal to
print (t >= u)

#Less than or equal to
print (t <= u)
```

```python
t = 1
u = 2

print (t < 5 and u < 10)

print(t < 5 or u < 4)

print (not(t < 5 and u < 10))
```
