```python
class Car:
    car_type = "Sedan"

    def __init__(self, name, mileage):
        self.name = name
        self.mileage = mileage

    def description(self):
        # Added 'f' here to make it an f-string
        return f"The {self.name} car gives the mileage of {self.mileage}km/l"

    def max_speed(self, speed):
        # Added 'f' here as well
        return f"The {self.name} runs at the maximum speed of {speed}km/hr"

# Creating an instance (Object)
my_car = Car("Toyota", 18)

print(my_car.description())
print(my_car.max_speed(150))
```

```python
import math

def calculate_area():
    print("--- Shapes Area Calculator ---")
    shape = input("Choose: triangle, circle, square, or rectangle: ").lower()

    if shape == "triangle":
        b = float(input("Base: "))
        h = float(input("Height: "))
        area = 0.5 * b * h

    elif shape == "circle":
        r = float(input("Radius: "))
        area = math.pi * r**2

    elif shape == "square":
        s = float(input("Side: "))
        area = s**2

    elif shape == "rectangle":
        l = float(input("Length: "))
        w = float(input("Width: "))
        area = l * w

    else:
        print("Invalid shape selection.")
        return

    print(f"The area of the {shape} is: {area:.2f}")

calculate_area()
```
